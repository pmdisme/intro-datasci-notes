# Inferrential Statistics

## Law of Large numbers

```python
df.total_bill.mean()
```

```python
df.sample(5)
```

```python
df.sample(5).total_bill.mean()
```

## การกระจายตัวของสถิติของกลุ่มตัวอย่าง

```python
ls = []
for i in range(30):
  ls.append(df.sample(5).total_bill.mean())
sample_5 = pd.Series(ls)
sample_5
```

```python
ls = []
for i in range(30):
  ls.append(df.sample(50).total_bill.mean())
sample_50 = pd.Series(ls)
sample_50
```

# Standard error

```python
sample_5.mean()
```

```python
sample_50.mean()
```

```python
sample_50.std()
```

```python
sns.displot(sample_50, kind='kde', height=3)
```

## Confidence interval

```python
sample_50.quantile([0.05, 0.95])
# confidence interval at the 90% confidence level
```

```python
sample_5.quantile([0.05, 0.95])
```

```python
sample_50.quantile([0.025, 0.975])
```
