# Resample

```python
# index must be timestamp first
df_house.resample('M').price.mean()
```

```python
df_house.resample('Q').size()
# Q : Quater
# Q end in DEC
```

```python
df_house.resample('Q-MAY').size()
# Q end in MAY
```

```python
df_house.resample('2M').size()
# 2M (count as month 5-6, 7-8, 9-10)
```

# Period

```python
df_house.resample('M').size()
```

```python
df_house.resample('M').size().to_period()
```

# Diff

```python
monthly_sale = df_house.resample("M").size()
monthly_sale
```

```python
monthly_sale.diff(1)
```

```python
monthly_sale.diff(3)
```

```python
monthly_sale.diff(-1)
```

# Shift

```python
monthly_sale.shift(1)
```

```python
df_compare = pd.concat([monthly_sale, monthly_sale.shift(1)], axis=1)
df_compare.columns = ['monthly_sale', 'last_month_sale']
df_compare
```

```python
df_compare.monthly_sale / df_compare.last_month_sale
```

# Rolling window

```python
three_months_window = monthly_sale.rolling(3)
```

```python
for window in three_months_window:
  print(window)
```
