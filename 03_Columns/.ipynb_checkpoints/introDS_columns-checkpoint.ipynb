```python
import pandas as pd
```

```python
df = pd.read_csv('players.csv')
df
```

```python
df.position
```

```python
df.sort_values('position')
# sort by letters in position
```

# Creating categorical column

```python
pd.Categorical(df.position)
```

```python
df.position.astype('category')
```

# Ordered categorical

```python
pd.Categorical(df.position, ordered=True)
```

```python
pos_cat = pd.Categorical(df.position, ordered=True,
               categories=['goalkeeper', 'defender', 'midfielder', 'forward'])
```

```python
df['pos_cat'] = pos_cat
df
```

```python
df.sort_values('pos_cat')
```

# Handling Categorical Data

```python
df.pos_cat.cat.categories
# use cat.categories
```

```python
df.pos_cat.cat.add_categories('manager')
```

```python
df.pos_cat.cat.remove_categories('goalkeeper')
```

```python
df.pos_cat.cat.reorder_categories(
    ['forward', 'midfielder', 'defender', 'goalkeeper']
)
```

```python
df.pos_cat.cat.as_unordered()
```
